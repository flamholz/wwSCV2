```python
%matplotlib inline
import xml.etree.ElementTree as ET
import pandas as pd
import numpy as np
```

```python
tree = ET.parse('data/051820/multicomponentdata.xml')
root = tree.getroot()
```

```python
FAM_data = {}
ROX_data = {}
fluor_order = 'FAM,ROX'.split(',')
    
for sd in root.iter('SignalData'):
    well_idx = sd.get('WellIndex')
    

    for idx, cd in enumerate(sd.iter('CycleData')):
        raw_data = cd.text
        parsed = list(map(float, raw_data.strip('[]').split(',')))
        if idx == 0:
            FAM_data[well_idx] = parsed
        else:
            ROX_data[well_idx] = parsed
    
FAM_df = pd.DataFrame(FAM_data)
ROX_df = pd.DataFrame(ROX_data)
```

```python
FAM_df.plot()
```
